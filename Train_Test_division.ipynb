```python
import numpy as np
from sklearn.model_selection import train_test_split
X = np.arange(5)
y = np.arange(5)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
from sklearn.datasets import load_boston
data = load_boston()
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```
